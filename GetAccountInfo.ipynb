#### Add Account Info

```python
def get_account_info(handle):
    c = Config()
    c.Username = handle
    c.Pandas = True
    run.Lookup(c)
    info_df = storage.panda.User_df
    del c
    return info_df['bio'], info_df['following'], info_df['followers']
```

```python
data['Bio'] = "None"
data['Following'] = 0
data['Followers'] = 0
for i in range(data.shape[0]):
    handle = data['Handle'][i]
    if handle not in ("NO_HANDLE", "SUSPENDED"):
        print(handle)
        bio, following, followers = get_account_info(handle)
        data.iloc[i, 9] = bio
        data.iloc[i,10] = following
        data.iloc[i,11] = followers
```
